```python
import pandas as pd
```

```python
df_clientes = pd.read_csv('customer_data.csv')
```

```python
df_clientes.head()
```

```python
df_ventas = pd.read_csv('sales_data.csv')
```

```python
df_ventas.head()
```

```python
df_final = pd.merge(df_ventas, df_clientes, on='customer_id', how='left')
```

```python
df_final.head()
```

```python
df_final.info()
```

```python
#Calculo de la Media
```

```python
media_edad = df_final['age'].mean()
```

```python
media_edad
```

```python
#Relleno de los valores nulos     para que no de float :   media_edad = int(df_final["age"].mean())    df_final["age"] = df_final["age"].fillna(media_edad)
```

```python
df_final['age'] = df_final['age'].fillna(media_edad)
```

```python
df_final['age'].isnull().sum()
```

```python
#Convertir la columna de fecha a formato datetime
```

```python
df_final['invoice_date'] = pd.to_datetime(df_final['invoice_date'],format='%d-%m-%Y',errors='coerce'  # Si alguna fecha es inválida, la convierte a NaT (Not a Time)
)
```

```python
# Comprobar la conversion
```

```python
df_final.info()
```

```python
# Creaciòn de grupos etarios, definiendo limites y etiquetas de cada uno
```

```python
bins = [18, 25, 36, 50, 65, 100]
labels = ['18-24', '25-35', '36-49', '50-64', '65+']
```

```python
df_final['age_group'] = pd.cut(
    df_final['age'],
    bins=bins,
    labels=labels,
    right=False,
    include_lowest=True
)
```

```python
df_final[['age', 'age_group']].head(10)
```

```python
df_final['age_group'].value_counts()
```

```python
#Métodos de pago grupo etario 25-35
```

```python
metodos_pago_25_35 = (
    df_final.loc[df_final['age_group'] == '25-35', 'payment_method']
    .value_counts()
)
```

```python
print(metodos_pago_25_35)
```

```python
#Métodos de pago grupo etario Mujeres
```

```python
metodos_pago_mujeres = (
    df_final.loc[df_final['gender'] == 'Female', 'payment_method']
    .value_counts()
)
```

```python
print(metodos_pago_mujeres)
```

```python
# 2. Configuración de la conexión
```

```python
import pandas as pd
from sqlalchemy import create_engine
import getpass # Para ingresar la contraseña de forma segura

# --- Variables de Conexión ---
USUARIO = 'root'      
PASSWORD = getpass.getpass('root: ') # Ingresa tu contraseña de forma segura
HOST = 'localhost'                # O la IP de tu servidor MySQL
PUERTO = 3306                     # Puerto estándar de MySQL
BASE_DE_DATOS = 'abp_datos_limpios' # Debe existir previamente en MySQL

# Crear la cadena de conexión (URL de SQLAlchemy)
# El formato es: 'dialect+driver://user:password@host:port/database'
CONN_STRING = f"mysql+pymysql://{USUARIO}:{PASSWORD}@{HOST}:{PUERTO}/{BASE_DE_DATOS}"
# Crear el motor de conexión
engine = create_engine(CONN_STRING)

print("Conexión con el motor SQLAlchemy creada.")
```

```python
# Carga del DataFrame a MySQL
```

```python
NOMBRE_TABLA = 'proyecto' # Nombre que tendrá la tabla en MySQL

try:
    df_final.to_sql(
        name=NOMBRE_TABLA, 
        con=engine, 
        if_exists='replace', # 'replace' borra y recrea la tabla si ya existe
        index=False,         # No carga el índice de Pandas como una columna
            )
    print(f"✅ ¡Carga exitosa! Datos cargados a la tabla '{NOMBRE_TABLA}' en MySQL.")
    
except Exception as e:
    print(f"❌ Error al cargar los datos: {e}")
```
